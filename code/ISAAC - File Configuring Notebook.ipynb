```python
## rename files in folders
import os

main_folder = 'New Video Frames'

for subfolder in os.listdir(main_folder):
    subfolder_path = os.path.join(main_folder, subfolder)
    
    # Check if it's a directory (subfolder)
    if os.path.isdir(subfolder_path):
        # Loop through each file in the subfolder
        for file_name in os.listdir(subfolder_path):
            if file_name.startswith("frame_") and file_name.endswith(".jpg"):
                # New file name without the "frame_" prefix
                new_name = file_name.replace("frame_000", "")
                
                if new_name.startswith("0") and len(new_name) == 6:  # Checking for format "0X.jpg" (total 6 chars)
                    new_name = new_name[1:]  # Removes the leading "0"
                
                # Full paths for the old and new file names
                old_file_path = os.path.join(subfolder_path, file_name)
                new_file_path = os.path.join(subfolder_path, new_name)
                
                # Rename the file
                os.rename(old_file_path, new_file_path)
                print(f"Renamed '{file_name}' to '{new_name}' in '{subfolder}'")
```
